```
import pandas as pd
rating=pd.read_excel('UserRatings (1).xlsx')
book=pd.read_excel('Books and author.xlsx')
```

```
print(book)
```

```
print(rating)
```

```
merged_df=pd.merge(rating,book,on='ISBN')
merged_df.info()
merged_df.drop(['Author','ISBN'],axis=1,inplace=True)
merged_df.info()
```

```
user_book_matrix=merged_df.pivot_table(values='Rating',index='User-ID',columns='Title')
user_book_matrix
```

```
corr_mat=user_book_matrix.corr()
corr_mat
```

```
print("THE ITEM-ITEM SIMILARTY IS CALCULATED USING PEARSON CORRELATION.")
print('They are as follows\n')
count=1
for user in book['Title']:
  lst=[]
  similar_movies=corr_mat[user].sort_values(ascending=False)
  similar_movies=similar_movies[similar_movies>0.7]
  for col in similar_movies.index[1:]:
    if col!=user:        
      temp=user_book_matrix.loc[:,[col,user]]
      temp.dropna(axis=0,inplace=True)
      if len(temp)>=3:
        #print(temp)
        lst.append(col)
  print('{} - The Books Similar to "{}" are {}'.format(count,user,lst))
  print('')
  count+=1
```
